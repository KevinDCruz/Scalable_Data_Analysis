```python
#Question 4
## Part A

LU2 = property1[property1['LU'] == 'R1']
#LU2
median_taxrate = (LU2['GROSS_TAX']/LU2['AV_TOTAL']).median(axis=0)


#Final Result of Part 4a
print ("Median Tax Rate is : " , median_taxrate)    
print("The median tax rate is in percentage. However, the Boston Property Residential Tax Rate is $10.59. Tax Rate is calculated in per thousand dollars as  per the website. Hence, (1.058*1000)/100=10.59. The answer obtained is correct.")
```

```python
#Question 4B

## Taking Frequency of Street in the Data
street_counts = LU2['ST_NAME'].value_counts()

## Filtering out the Street which has Single Family Apartments >20  
Street_R1_20 = street_counts[street_counts>20]

## Dropping the Null values
Street_R1_20 = Street_R1_20.dropna()

## Filtering out the Streets with less than 20 single resident apartments

R1_20  = LU2[LU2.ST_NAME.apply(lambda x : x in Street_R1_20)]

mean = R1_20.groupby('ST_NAME')['GROSS_TAX'].mean().reset_index()
mean
```

```python
#street = property1['AV_TOTAL'].argmax()
#highestvalue = property1.iloc[max]

street = mean.iloc[mean['GROSS_TAX'].argmax()]

#Final Result of 4B
street                     
```
